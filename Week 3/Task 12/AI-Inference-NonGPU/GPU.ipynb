```python
import time
from transformers import pipeline

# Start timer
start = time.time()

# Load model
classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

# Input text
text = "AI inference comparison between GPU and CPU"

# Run inference
result = classifier(text)

# End timer
end = time.time()

# Output
print(result)
print("GPU Time:", end - start, "seconds")
```
