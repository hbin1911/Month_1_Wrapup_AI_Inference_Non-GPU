```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

result = classifier("I love working on AI projects")

print(result)
```

```python
from transformers import pipeline

ner = pipeline("ner", aggregation_strategy="simple")

text = "Rajas Mulik is working on AI project in Mumbai on 12/03/2024"

output = ner(text)

print(output)
```
