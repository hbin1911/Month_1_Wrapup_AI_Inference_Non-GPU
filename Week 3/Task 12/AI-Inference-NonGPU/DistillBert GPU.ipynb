```python
from transformers import pipeline
```

```python
classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
text = "AI inference in non-GPU environments is very interesting"
```

```python
result = classifier(text)

print(result)
```
