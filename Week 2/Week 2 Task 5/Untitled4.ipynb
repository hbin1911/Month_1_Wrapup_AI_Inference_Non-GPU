```python
# List comprehension
squares = [x*x for x in range(5)]
print(squares)

# Function
def greet(name):
    return f"Hello {name}"

print(greet("Rajas"))

# Class
class Student:
    def __init__(self, name):
        self.name = name

s = Student("Rajas")
print(s.name)

# File I/O
with open("test.txt", "w") as f:
    f.write("Python Basics Done")

with open("test.txt", "r") as f:
    print(f.read())
```
